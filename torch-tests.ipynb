```python
import torch
import torch.nn as nn
import torch.nn.functional as F

vocab = ["king", "queen", "man", "woman"]
word_to_ix = {word: i for i, word in enumerate(vocab)}
ix_to_word = {i: word for word, i in word_to_ix.items()}
print(word_to_ix)
print(ix_to_word)
```

```python
training_pairs = [
    ("king", "man"),
    ("queen", "woman"),
    ("king", "queen"),
    ("man", "woman"),
]

data = [(word_to_ix[w1], word_to_ix[w2]) for w1, w2 in training_pairs]
print(data)
```

```python
embedding_dim = 2  # 2D so we can visualize easily

class SimpleEmbedModel(nn.Module):
    def __init__(self, vocab_size, emb_dim):
        super().__init__()
        self.embeddings = nn.Embedding(vocab_size, emb_dim)

    def forward(self, input_id):
        return self.embeddings(input_id)
```

```python
model = SimpleEmbedModel(vocab_size=len(vocab), emb_dim=embedding_dim)
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)

for epoch in range(1000):
    total_loss = 0
    for input_ix, target_ix in data:
        input_tensor = torch.tensor([input_ix], dtype=torch.long)
        target_tensor = torch.tensor([target_ix], dtype=torch.long)
        print(input_tensor)
        print(target_tensor)
        

        # Get embedding and predict target via dot product
        input_vec = model(input_tensor)
        emb = model.embeddings.weight.data
        for i in range(len(vocab)):
            print(f"{vocab[i]}: {emb[i].numpy()}")
        target_vec = model(target_tensor)
        emb = model.embeddings.weight.data
        for i in range(len(vocab)):
            print(f"{vocab[i]}: {emb[i].numpy()}")
        score = F.cosine_similarity(input_vec, target_vec)
        print(score)

        # Want cosine similarity to be 1 (i.e., same direction)
        loss = (1 - score)**2
        total_loss += loss.item()

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    if epoch % 200 == 0:
        print(f"Epoch {epoch}, Loss: {total_loss:.4f}")
```

```python
emb = model.embeddings.weight.data

for i in range(len(vocab)):
    print(f"{vocab[i]}: {emb[i].numpy()}")
```

```python
import matplotlib.pyplot as plt

for word, i in word_to_ix.items():
    vec = emb[i]
    plt.scatter(vec[0], vec[1])
    plt.text(vec[0] + 0.01, vec[1] + 0.01, word)

plt.title("2D Word Embeddings after Training")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

words = ['king', 'queen', 'man', 'woman']
vectors = [emb[word_to_ix[w]].numpy() for w in words]

for word, vec in zip(words, vectors):
    plt.scatter(vec[0], vec[1])
    plt.text(vec[0] + 0.02, vec[1] + 0.02, word)

plt.grid(True)
plt.title("Trained 2D Word Embeddings")
plt.show()
```
